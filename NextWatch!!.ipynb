```python
import numpy as np 
import pandas as pd 
import ast 
import nltk
```

```python
df_anime=pd.read_csv('AnimeList.csv')
df_movies=pd.read_csv('Movies.csv')
df_Tvshows=pd.read_csv('Tvshows.csv')



#1st anime---------------

df_anime.head()
```

```python
##rename the 'genre'------> 'genres'
##rename the 'anime_id'------> 'id'
df_anime.rename(columns={'genre': 'genres'}, inplace=True)
df_anime.rename(columns={'anime_id': 'id'}, inplace=True)
df_anime.head()
```

```python
df_anime= df_anime[['id','title','episodes','studio','genres']]
df_anime.head()
```

```python
df_anime.isnull().sum()
```

```python
df_anime.dropna(inplace=True)
```

```python
df_anime.duplicated().sum()
```

```python
##split and remove spaces from the studio,genres column
df_anime['studio']=df_anime['studio'].apply(lambda x: x.split())
df_anime['genres']=df_anime['genres'].apply(lambda x: x.split())
df_anime['studio']=df_anime['studio'].apply(lambda x: [i.replace(" ","")for i in x])
df_anime['genres']=df_anime['genres'].apply(lambda x: [i.replace(" ","")for i in x])

```

```python
#2nd : movies------------------------------------------------------

df_movies.head()
```

```python
df_movies=df_movies[['genres','id','keywords','title','overview']]
df_movies.head()
```

```python
df_movies.isnull().sum()
```

```python
df_movies.dropna(inplace=True)
df_movies.duplicated().sum()
```

```python
##need to fixed the genres and overview columns

def convert(obj):
    L=[]

    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

```python
df_movies['genres']=df_movies['genres'].apply(convert)
df_movies['keywords']=df_movies['keywords'].apply(convert)

df_movies.head()
```

```python
##converting a full string(overview column) into dict of string
df_movies['overview']= df_movies['overview'].apply(lambda x: x.split())

#remove spaces from those string type columns
df_movies['genres']= df_movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
df_movies['keywords']= df_movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
df_movies['overview']= df_movies['overview'].apply(lambda x :[i.replace(" ","") for i in x])
```

```python
#3rd : series-------------------------------------------------

df_Tvshows.head()
```

```python
df_Tvshows=df_Tvshows[['id','genre_ids','name','overview']]
df_Tvshows.head()
```

```python
df_Tvshows.isnull().sum()
```

```python
df_Tvshows.dropna(inplace=True)
df_Tvshows.duplicated().sum()
df_Tvshows.drop_duplicates(inplace=True)
```

```python
# 1. Define the TMDB TV Genre Mapping
genre_map = {
    10759: "Action&Adventure", 16: "Animation", 35: "Comedy", 80: "Crime", 
    99: "Documentary", 18: "Drama", 10751: "Family", 10762: "Kids", 
    9648: "Mystery", 10763: "News", 10764: "Reality", 10765: "SciFi&Fantasy", 
    10766: "Soap", 10767: "Talk", 10768: "War&Politics", 37: "Western"
}

df_Tvshows['genres']=df_Tvshows['genre_ids'].apply(lambda x: [genre_map[i] for i in x if i in genre_map])
##rename the 'name'------> 'title'
df_Tvshows.rename(columns={'name': 'title'}, inplace=True)
```

```python
##converting a full string(overview column) into dict of string
df_Tvshows['overview']=df_Tvshows['overview'].apply(lambda x : x.split())
#remove spaces from sentence
df_Tvshows['overview']=df_Tvshows['overview'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
##now merge these 3 datasets datas into a single column---------> tags

df_movies['tags']= df_movies['genres']+df_movies['keywords']+df_movies['overview']

df_Tvshows['tags']= df_Tvshows['genres']+df_Tvshows['overview']

df_anime['tags']= df_anime['genres']+df_anime['studio']
```

```python
##now add a new column content_type for these 3 datasets

df_anime['content_type']='Anime'
df_movies['content_type']='Movie'
df_Tvshows['content_type']='Tvshows'

##now the needed column should only exist

df_newanime= df_anime[['id','title','tags','content_type']]
df_newmovies= df_movies[['id','title','tags','content_type']]
df_newTvshows= df_Tvshows[['id','title','tags','content_type']]
```

```python
##finally merge these 3 datasets and creating the model

main_df= pd.concat([df_newanime, df_newmovies, df_newTvshows], ignore_index=True)

# Convert the 'tags' list back into a single string for Text Vectorization
main_df['tags']= main_df['tags'].apply(lambda x:" ".join(x))
main_df['tags']= main_df['tags'].apply(lambda x: x.lower())

main_df.tail()
```

```python
## Text vectorization
from sklearn.feature_extraction.text import CountVectorizer

#We want the top 5000 most frequent words to avoid "noise"
# We also remove stop words like "the", "a", "is"
cv= CountVectorizer(max_features=5000, stop_words='english')
```

```python
from nltk.stem.porter import PorterStemmer ##to remove the same tense words in this column
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)


main_df['tags'].apply(stem)
main_df.tail()
```

```python
## vectorize my data
vectors= cv.fit_transform(main_df['tags']).toarray()
vectors
```

```python
#now use cosine similarity to find the movies from user interest

from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
```

```python

import joblib


# 4. Save to files (THE BETTER BRIDGE TO STREAMLIT)
# It's common to keep the .pkl extension, or use .joblib
joblib.dump(main_df, 'main_df.joblib')
joblib.dump(similarity, 'similarity.joblib')

```
